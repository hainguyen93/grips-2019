```python
schedules = []
f = open("output.sol", "r")
for line in f.readlines():
    if "var_x_" in line and '1' in line.split(' ')[1]:
        schedules.append(line)
```

Run through schedules to count the number of inspectors in the schedule

```python
inspectors = set()
for line in schedules:
    inspectors.add(line.split('#')[0][-1:])
```

```python
inspectors
```

For each inspector, extract their daily schedule

```python
for k in inspectors:
    for line in schedules:
        if k in line.split("#")[0][-1:]:
            depart = line.split('_')[2]
            arrive = line.split('_')[3]
            print("depart from {} at {} and arrive to {} at {}".format(depart.split('@')[0],depart.split('@')[1],arrive.split('@')[0], arrive.split('@')[1] ))
```

I need to account for sinks and sources. So I will collect those first

```python
import pandas as pd
S = []
for k in inspectors:
    df = pd.DataFrame(columns=['inspector_ID', 'start_station','start_time','end_station','end_time'])
    S.append(df)
```
